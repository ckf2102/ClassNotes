**3. Define cool/useful/funny as the features, and stars as the response.**

```python
feature_cols = ['cool', 'useful', 'funny']

yelp_X = data[feature_cols]
yelp_y = data.stars
```

**4. Fit a linear regression model and interpret the coefficients. Do the coefficients make intuitive sense to you? Explore the Yelp website to see if you detect similar trends.**

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics

linreg = LinearRegression()
linreg.fit(yelp_X, yelp_y)

print linreg.intercept_ 

zip(feature_cols, linreg.coef_)
```

Using multiple regression:
- The slope of "cool" = 0.27; this means that assuming "useful" and "funny" are constant, an increase in one cool vote is associated with an increase in 0.27 star rating
- The slope of "useful" = -0.15; this means that assuming "cool" and "funny" are constant, an increase in one useful vote is associated with a decrease in 0.15 star rating
- The slope of "funny" = -0.14; this means that assuming "cool" and "useful" are constant, an increase in one funny vote is associated with a decrease in 0.14 star rating

This makes some sense to me-- I tend to find that the lower rating reviews are sarcastic and funny. Reviews that are voted useful could go either way, so the slope being close to 0 makes sense to me. 


**5. Evaluate the model by splitting it into training and testing sets and computing the RMSE. Does the RMSE make intuitive sense to you?**

```python
from sklearn.cross_validation import train_test_split

def train_test_rmse(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    y_pred = linreg.predict(X_test)
    return np.sqrt(metrics.mean_squared_error(y_test, y_pred))

train_test_rmse(yelp_X, yelp_y)
```

RMSE makes more sense than other measures of error because it's relative to the response values.
